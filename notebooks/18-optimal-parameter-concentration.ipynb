# MAXCUT Landscape Analysis

```python
import numpy as np
import matplotlib.pyplot as plt

from qaoa_vrp.solutions.solutions import *
```

```python
n_values = np.arange(1, 51)  # System sizes from 1 to 50

def plot_beta_gamma(n_layers):
    nrows = (n_layers + 1) // 2
    ncols = 2 if n_layers > 1 else 1

    fig, axs = plt.subplots(nrows, ncols, figsize=(15, 5 * nrows), squeeze=False)
    fig.suptitle('Beta and Gamma vs System size for Different Layer Counts')

    for i in range(1, n_layers + 1):
        beta_values = []
        gamma_values = []

        for n in n_values:
            params = get_analytical_parameters(n, n_layers=n_layers)
            beta_values.append(params[i - 1])  # beta_i
            gamma_values.append(params[n_layers + i - 1])  # gamma_i

        ax = axs[(i-1) // ncols, (i-1) % ncols]

        # Plot beta and gamma values for the ith layer
        ax.plot(n_values, beta_values, '-', label=fr'$\beta_{i}$')
        ax.plot(n_values, gamma_values, '-', label=fr'$\gamma_{i}')
        ax.scatter(n_values, beta_values)
        ax.scatter(n_values, gamma_values)

        ax.set_xlabel('System size ($n$)')
        ax.set_ylabel('Parameter Value')
        ax.set_title(f'Layer {i}')
        ax.grid(True)
        ax.legend()

    plt.tight_layout()
    plt.show()
```

```python
plot_beta_gamma(n_layers=1)
```

```python
plot_beta_gamma(n_layers=2)
```

```python
plot_beta_gamma(n_layers=3)
```
