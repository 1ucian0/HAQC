```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d_results = pd.read_csv("../stupid-seaborn-data.csv")
```

```python
d_results_2 = d_results.copy(deep=True)
d_results_2["layer"] = 3
d_results_3 = d_results.copy(deep=True)
d_results_3["layer"] = 4
d_results = pd.concat([d_results, d_results_2, d_results_3])
```

```python
g = sns.relplot(data=d_results, x="total_evals", y="value", col="layer", kind="line", col_wrap=2)
axes = g.axes.flatten()
for ax in axes:
            ax.axhline(
                -180, ls='--', linewidth=3, color='grey'
            )
            ax.set_xlabel("Function Evals")
plt.show()            
```
